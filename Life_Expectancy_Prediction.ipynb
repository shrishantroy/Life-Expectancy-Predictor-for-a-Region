```python
df = pd.get_dummies(df,columns = ['Status'])
df.head()
```

```python
df['Status_Developed'] = df['Status_Developed'].astype(int)
df['Status_Developing'] = df['Status_Developing'].astype(int)
df.head()
```

```python
y = df['Life expectancy ']
x = df.drop(columns = 'Life expectancy ')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
x_train = x_train.apply(lambda x : x.fillna(x.mean()),axis = 0)
x_test = x_test.apply(lambda x : x.fillna(x.mean()),axis = 0)
```

```python
x_train.isnull().sum()
```

```python
x_test.isnull().sum()
```

```python
y_train = y_train.fillna(y_train.mean())
y_test = y_test.fillna(y_test.mean())
```

```python
from xgboost import XGBRegressor
xg = XGBRegressor(objective = 'reg:squarederror',learning_rate = 0.05,max_depth = 5, n_estimators = 200)
xg.fit(x_train,y_train)
```

```python
print("R2 score :",xg.score(x_test,y_test))
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from math import sqrt

y_pred = xg.predict(x_test)
rmse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

k = x_train.shape[1]
n = x_train.shape[0]
adj_r2 = 1 - ((1-r2)*(n-1)/(n-k-1))

print("RMSE :",rmse)
print("MAE :",mae)
print("R2 :",r2)
print("Adjusted R2 :",adj_r2)
```

```python
MRE = (np.mean(abs(y_test - y_pred)/y_test))*100
print(f"Mean relative error: {round(MRE,3)} %",)
```

```python
plt.scatter(y_test,y_pred)
plt.title("Prdicted vs actual")
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.grid()
plt.show()
```
